```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('data/corona_tested_individuals_ver_006.english.csv')
print(df.shape)
df.head()
```

```python
df = df.loc[df['corona_result'].ne('other')]
df = df.loc[df['age_60_and_above'].ne('None')]
df = df.loc[df['gender'].ne('None')]
print(df['corona_result'].unique())
print(df['age_60_and_above'].unique())
print(df['gender'].unique())
```

```python
labelencoder = LabelEncoder()
df['corona_result'] = labelencoder.fit_transform(df['corona_result'])
df['age_60_and_above'] = labelencoder.fit_transform(df['age_60_and_above'])
df['gender'] = labelencoder.fit_transform(df['gender'])
df.loc[df['cough'] == 'None', "cough"] = 0
df.loc[df['fever'] == 'None', "fever"] = 0
df.loc[df['test_indication'] == 'Contact with confirmed', "test_indication"] = 1
df.loc[df['test_indication'] == 'Other', "test_indication"] = 0
df.loc[df['test_indication'] == 'Abroad', "test_indication"] = 0
df.head()
```

```python
df.to_csv('data/corona_tested_individuals_ver_006.english_cleaned.csv')
```
